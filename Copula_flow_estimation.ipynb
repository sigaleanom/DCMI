```python
from aux_functions import *
from spline_flow import SplineFlow
from affine_ar_flow import AffineARFlow
```

```python
import pandas as pd
```

```python
def load_data(path, distr = normal):
    df = pd.read_csv(path).iloc[:, 1:]
    uniform_scale = torch.FloatTensor(df.values).to(device)
    return uniform_scale, distr.icdf(uniform_scale)
```

```python
def plot_scales(original, rescaled):
    original = original.to("cpu").detach().numpy()
    rescaled = rescaled.to("cpu").detach().numpy()
    
    fig, ax = plt.subplots(1, 2, figsize = (12, 5))
    ax[0].scatter(*original.T, s=10, alpha=0.7)
    ax[0].set_title("Data in the uniform Scale")

    ax[1].scatter(*rescaled.T, s=10, alpha=0.7)
    ax[1].set_title("Data in the Gaussian Scale")

    plt.show()
    
```

# Clayton Copula

```python
clayton, clayton_rescaled = load_data("./Clayton.csv")
plot_scales(clayton, clayton_rescaled)
```

## AR Flow

```python
clayton_flow_ar = AffineARFlow(2, device, num_flows = 9)
train_flow(clayton_flow_ar, clayton_rescaled)
visualize_flow(clayton_flow_ar, clayton_rescaled)
visualize_flow(clayton_flow_ar, clayton_rescaled, copula_scale=True)
```

## Spline Flow

```python
clayton_flow_spline = SplineFlow(2, device, num_flows = 4)
train_flow(clayton_flow_spline, clayton_rescaled)
visualize_flow(clayton_flow_spline, clayton_rescaled)
visualize_flow(clayton_flow_spline, clayton_rescaled, copula_scale=True)
```
